```python
import pandas as pd

df = pd.read_csv("DataCoSupplyChainDataset.csv", encoding="latin1")

# Genel boş veri özeti
print("=== BOŞ VERİ ANALİZİ ===\n")

toplam_bos = df.isnull().sum().sum()
print(f"Toplam boş hücre sayısı: {toplam_bos}")
print(f"Toplam hücre sayısı: {df.size}")
print(f"Boş veri oranı: %{toplam_bos / df.size * 100:.2f}\n")

# Sütun bazında boş veri
bos_sutunlar = df.isnull().sum()
bos_sutunlar = bos_sutunlar[bos_sutunlar > 0].sort_values(ascending=False)

if len(bos_sutunlar) == 0:
    print("Hiç boş veri yok!")
else:
    print(f"Boş veri içeren sütun sayısı: {len(bos_sutunlar)}\n")
    print(f"{'Sütun':<30} {'Boş Sayı':>10} {'Oran':>10}")
    print("-" * 52)
    for sutun, sayi in bos_sutunlar.items():
        oran = sayi / len(df) * 100
        print(f"{sutun:<30} {sayi:>10} {oran:>9.2f}%")

# Satır bazında özet
satirlarda_bos = df.isnull().any(axis=1).sum()
print(f"\nEn az 1 boş değer içeren satır sayısı: {satirlarda_bos} ({satirlarda_bos/len(df)*100:.2f}%)")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("DataCoSupplyChainDataset-Cleaned.csv", encoding="latin1")

# 1. SÜTUN İSMİ TEMİZLİĞİ (BOM karakteri düzelt)
df.rename(columns={"ï»¿Type": "Type"}, inplace=True)

# 2. DUPLICATE KONTROLÜ
print(f"Tekrar eden satır: {df.duplicated().sum()}")

# 3. VERİ TİPLERİ - Tarih sütunlarını düzelt
df["order date (DateOrders)"] = pd.to_datetime(df["order date (DateOrders)"])
df["shipping date (DateOrders)"] = pd.to_datetime(df["shipping date (DateOrders)"])
print("\nVeri tipleri:\n", df.dtypes)

# 4. MANTIKSAL TUTARLILIK
# Gerçek gün > planlanan gün mı? (geç teslimat)
gec = (df["Days for shipping (real)"] > df["Days for shipment (scheduled)"]).sum()
print(f"\nGeç gönderilen sipariş sayısı: {gec}")

# Negatif olmaması gerekenler
for col in ["Sales", "Order Item Quantity", "Product Price", "Order Item Total"]:
    neg = (df[col] < 0).sum()
    if neg > 0:
        print(f"⚠️  {col} sütununda {neg} negatif değer var!")

# 5. SIFIR DEĞERLERİ
print("\nSayısal sütunlarda sıfır değerler:")
sayi_cols = df.select_dtypes(include="number").columns
sifirlar = (df[sayi_cols] == 0).sum()
print(sifirlar[sifirlar > 0])

# 6. KATEGORİK TUTARSIZLIK (büyük/küçük harf, boşluk)
kategorik = ["Type", "Delivery Status", "Customer Segment", "Market", 
             "Order Status", "Shipping Mode", "Department Name", "Category Name"]
print("\nKategorik sütun unique değerleri:")
for col in kategorik:
    print(f"\n{col}: {df[col].unique()}")

# 7. OUTLIER - IQR yöntemi
print("\nAykırı değer analizi:")
for col in ["Benefit per order", "Sales per customer", "Order Item Profit Ratio", 
            "Order Profit Per Order", "Sales"]:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outlier = ((df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)).sum()
    print(f"{col}: {outlier} aykırı değer")

# 8. GİZLİLİK - Hassas sütunları düşür
hassas = ["Customer Email", "Customer Password", "Customer Street", "Product Image"]
df.drop(columns=hassas, inplace=True)
print(f"\nHassas sütunlar silindi: {hassas}")

# 9. GEREKSIZ SÜTUNLAR
# Product Status - kaç unique değer var?
print(f"\nProduct Status unique: {df['Product Status'].unique()}")
# Eğer tek değer varsa anlamsız, sil

# 10. TEMİZ VERİYİ KAYDET
df.to_csv("temiz_veri.csv", index=False, encoding="utf-8-sig")
print("\n✅ Temiz veri kaydedildi!")
```

```python
# 1. ANLAMSIZ SÜTUNU SİL
df.drop(columns=["Product Status"], inplace=True)
print("Product Status silindi.")

# 2. KATEGORİK SÜTUNLARDA TRAILING SPACE TEMİZLE
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].str.strip()
print("Trailing space temizlendi.")

# 3. AYKIRI DEĞERLERİ İNCELE (silmeden önce bak)
print("\nBenefit per order istatistik:")
print(df["Benefit per order"].describe())
print("\nNegatif kar olan sipariş sayısı:", (df["Benefit per order"] < 0).sum())
print("Çok yüksek kar (>1000) sipariş sayısı:", (df["Benefit per order"] > 1000).sum())

# 4. ORDER DATE ZAMANSAL KONTROL
df["shipping_delay"] = (df["shipping date (DateOrders)"] - df["order date (DateOrders)"]).dt.days
print("\nTeslimat süresi istatistik:")
print(df["shipping_delay"].describe())
# Negatif olan var mı? (kargo sipariş öncesi gönderilmiş olamaz)
print(f"Negatif teslimat süresi: {(df['shipping_delay'] < 0).sum()}")

# 5. GEÇ TESLİMAT TUTARLILIĞI
# Late_delivery_risk ile Delivery Status uyuşuyor mu?
print("\nDelivery Status vs Late_delivery_risk çapraz kontrol:")
print(pd.crosstab(df["Delivery Status"], df["Late_delivery_risk"]))

# 6. ORDER STATUS - CANCELED siparişlerde satış sıfır mı?
print("\nİptal siparişlerde satış var mı?")
canceled = df[df["Order Status"] == "CANCELED"]
print(canceled[["Sales", "Order Item Total", "Benefit per order"]].describe())

# 7. TEMİZ VERİYİ TEKRAR KAYDET
df.to_csv("temiz_veri_v2.csv", index=False, encoding="utf-8-sig")
print("\n✅ temiz_veri_v2.csv kaydedildi!")
```

```python
# 1. NEGATİF KARI FLAG'LE (silme, işaretle)
df["negative_profit_flag"] = (df["Benefit per order"] < 0).astype(int)
print(f"Negatif kar flag'lendi: {df['negative_profit_flag'].sum()} sipariş")

# 2. İPTAL SİPARİŞLERİ FLAG'LE
df["is_canceled"] = (df["Order Status"] == "CANCELED").astype(int)
print(f"İptal sipariş flag'lendi: {df['is_canceled'].sum()} sipariş")

# 3. ANALİZ İÇİN TEMİZ SUBSET (iptal ve fraud hariç)
temiz_df = df[~df["Order Status"].isin(["CANCELED", "SUSPECTED_FRAUD"])]
print(f"\nAnaliz için geçerli sipariş sayısı: {len(temiz_df)}")
print(f"Çıkarılan sipariş sayısı: {len(df) - len(temiz_df)}")

# 4. SUSPECTED_FRAUD KONTROL
fraud = df[df["Order Status"] == "SUSPECTED_FRAUD"]
print(f"\nSuspected Fraud sipariş sayısı: {len(fraud)}")
print(fraud[["Sales", "Benefit per order", "Market", "Customer Segment"]].describe())

# 5. SON ÖZET
print("\n=== VERİ SETİ ÖZET ===")
print(f"Toplam sipariş: {len(df)}")
print(f"Geçerli sipariş (analiz için): {len(temiz_df)}")
print(f"İptal: {(df['Order Status'] == 'CANCELED').sum()}")
print(f"Suspected Fraud: {(df['Order Status'] == 'SUSPECTED_FRAUD').sum()}")
print(f"Negatif karlı: {df['negative_profit_flag'].sum()}")
print(f"Geç teslimat: {(df['Delivery Status'] == 'Late delivery').sum()}")
print(f"Geç teslimat oranı: %{(df['Delivery Status'] == 'Late delivery').sum() / len(df) * 100:.1f}")

# 6. FINAL TEMİZ VERİYİ KAYDET
df.to_csv("temiz_veri_final.csv", index=False, encoding="utf-8-sig")
temiz_df.to_csv("analiz_veri.csv", index=False, encoding="utf-8-sig")
print("\n✅ temiz_veri_final.csv (tüm veri, flag'li)")
print("✅ analiz_veri.csv (iptal/fraud hariç, analize hazır)")
```
